### Insights from the data 

1. What is the Average Cooling Efficiency in the system for the recorded time?

```python
# Calculate the average cooling efficiency
average_efficiency = df['Cooling efficiency'].mean()
print('Average Cooling Efficiency:', average_efficiency)
```

2. What is the maximum cooling power and date?

```python
# Find the maximum cooling power and its date
max_cooling_power = df['Cooling power'].max()
max_power_date = df.loc[df['Cooling power'] == max_cooling_power, 'Date'].values[0]
print("Maximum Cooling Power:", max_cooling_power)
print("Date of Maximum Cooling Power:", max_power_date)
```

```python
# Plot a scatter plot of cooling power and motor power
plt.figure(figsize = [10, 5])
plt.scatter(df['Cooling power'], df['Motor power W'], alpha = 0.5)
plt.xlabel('Cooling Power')
plt.ylabel('Motor Power')
plt.title('Cooling Power vs. Motor Power')
plt.show()
```

```python
# Calculate the correlation between cooling power and motor power
correlation = df['Cooling power'].corr(df['Motor power W'])
correlation 
```

There is a strong correlation between Cooling Power and Motor Power. 

3. How does the cooling efficiency vary over time?

```python
# Plot a line graph of cooling efficiency over time
plt.figure(figsize = [10, 5])
plt.plot(df['Time'], df['Cooling efficiency'])
plt.xlabel('Time')
plt.ylabel('Cooling Efficiency')
plt.title('Cooling Efficiency Over Time')
plt.show()
```

It can be seen that the cooling efficiency picks up after about each 45,000 seconds before it stabilizes, with a maximum cooling efficiency of 48Hz at 105,000 seconds.

```python
max_ce = df['Cooling efficiency'].max()
max_ce
```

4. What is the trend of cooling power over the recorded period?

```python
# To determine the trend of cooling power over the recorded period 
plt.figure(figsize = [10, 5])
df.plot(x='Time', y='Cooling power', kind='line')
plt.xlabel('Time')
plt.ylabel('Cooling Power')
plt.title('Cooling Power Trend Over Time')
plt.show()
```

There is an increase in cooling power after about each 45,000 seconds before a visible drop in cooling power  

```python
# Visualize the correlation matrix as a heatmap
plt.figure(figsize = [17, 10])
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# plotting a heatmap to view the relationship between Time and Vibration mm/s
variables = ['Time',' Vibration mm/s']
plt.figure(figsize = [8, 5])
sns.heatmap(df[variables].corr(), annot = True, fmt = '.3f',cmap = 'coolwarm', center = 0)
plt.show()
```

Time and Vibration has an inverse relationship in the hydraulic system. As time increases, the vibration mm/s decreases.

5. How does the motor power consumption change with time?

```python
# To analyze the change in motor power consumption over time
plt.figure(figsize = [15, 5])
df.plot(x='Time', y='Motor power W', kind='line')
plt.xlabel('Time')
plt.ylabel('Motor Power (W)')
plt.title('Motor Power Consumption Over Time')
plt.show()
```

There is a spike and drop in motor power for the first 40,000 seconds, after which the motor power gradually increases as time increases. 

6. Is there any correlation between temperature and vibration in the system?


```python
# To examine the correlation between temperature and vibration
plt.scatter(df['Temperature 1'], df[' Vibration mm/s'], alpha = 0.5)
#plt.xlim(20, 100)
plt.xlabel('Temperature 1')
plt.ylabel('Vibration (mm/s)')
plt.title('Temperature vs. Vibration')
plt.show()
```

```python
correlation = df['Temperature 1'].corr(df[' Vibration mm/s'])
correlation
```

There is a strong correlation between temperature and vibration in the system. As the temperature increases, the vibration increases gradually.

7. What is the relationship between temperature and pressure in the system?

```python
# To analyze the relationship between temperature and pressure
plt.figure(figsize = [8, 5])
plt.scatter(df['Temperature 1'], df['Pressure bar 1'])
#plt.xlim(20, 60)
plt.xlabel('Temperature 1')
plt.ylabel('Pressure (bar 1)')
plt.title('Temperature vs. Pressure')
plt.show()
```

```python
correlationT = df['Temperature 1'].corr(df['Pressure bar 1'])
correlationT
```

There is a strong negative relationship between Pressure and Temperature in the hydraulic system. As the temperature increases, the pressure reduces with few outliers .

8. Is there a relationship between volume flow and pressure in the system?

```python
# To explore the relationship between volume flow and pressure in the system
plt.figure(figsize = [15, 8])
plt.subplot(2, 2, 1)

sns.regplot(data = df, x= 'Volume flow l/min 1', y = 'Pressure bar 1')
plt.xlabel('Volume Flow (l/min 1)')
plt.ylabel('Pressure (bar 1)')
plt.title('Volume Flow 1 vs. Pressure 1')

plt.subplot(2, 2, 2)
#plt.figure(figsize = [8, 5])
sns.regplot(data = df, x= 'Volume flow l/min 2', y = 'Pressure bar 2')
plt.xlabel('Volume Flow (l/min 2)')
plt.ylabel('Pressure (bar 2)')
plt.title('Volume Flow 2 vs. Pressure 2')

plt.show()
```

```python
# Calculating the correlation between volume flow and pressure bar
correlation1 = df['Volume flow l/min 1'].corr(df['Pressure bar 1'])
correlation1
```

There is a weak correlation between volume flow 1 and pressure bar 1 but there is a strong correlation between Volumn flow 2 and Pressure bar 2.

9. How does the efficiency factor vary with time?

```python
# To analyze the variation of the efficiency factor over time
plt.figure(figsize = [10, 5])
df.plot(x='Time', y='Efficiency factor', kind='line')
plt.xlabel('Time')
plt.ylabel('Efficiency Factor')
plt.title('Efficiency Factor Variation Over Time')
plt.show()
```

For the first 420000 seconds, there is a significant drop and pickup in the efficeincy factor. As time increases, there is a gradual rise and fall in the efficiency factor of the system

10. How do changes in volume flow affect the efficiency and performance of the system?

```python
#To plot the volume flow against the efficiency factor
df.plot(x='Volume flow l/min 1', y='Efficiency factor', kind='line', label = 'Volume flow 1')
plt.xlabel('Volume Flow (l/min 1)')
plt.ylabel('Efficiency Factor')
plt.title('Volume Flow 1 vs. Efficiency Factor')

plt.show()
```

```python
#To plot the volume flow against the efficiency factor in heatmap
variables = ['Efficiency factor','Volume flow l/min 1']
plt.figure(figsize = [8, 5])
sns.heatmap(df[variables].corr(), annot = True, fmt = '.3f',cmap = 'coolwarm', center = 0)

plt.show()
```

There is a negative correlation between the efficiency factor and volume flow. Increased volume flow results in decrease in efficiency factor of the system.

```python
# plotting the volume flow against the cooling power
sns.regplot(data = df, x= 'Volume flow l/min 1', y='Cooling power', color = 'red', truncate = False, x_jitter = 0.3)
plt.xlabel('Volume Flow (l/min 1)')
plt.ylabel('Cooling Power')
plt.title('Volume Flow vs. Cooling Power')
```

Increase in volume flow leads to slight reduction in reduction in cooling power.
An increase in volume flow leads to decline in both the efficiency factor and cooling power.
